```python
import pandas as pd 
```

```python
# Load both combined data sets
sets = pd.read_csv('../Data/combined_set_data.csv')
full_data = pd.read_csv('../Data/combined_full_data.csv')
```

```python
# Identify the columns in full_data that are not in sets
static_columns = full_data.columns[~full_data.columns.isin(sets.columns)]

# Remove names that include "price" or 'release'
static_columns = static_columns[~static_columns.str.contains('price|release', case=False)]

# Remove views, viewsweek, upvotes, downvotes
static_columns = static_columns[~static_columns.str.contains('views|upvotes|downvotes', case=False)]

# static_columns

mask = full_data['time_pulled'] == max(full_data['time_pulled'])

# Add card_id, name, set_name, and set_code to static columns
choice_cols = static_columns.union(['card_id', 'name', 'set_name', 'set_code', 'rarity'])

static_data = full_data[mask][choice_cols]

static_data
```

```python
# Join static_data to sets
full = sets.merge(static_data, 
                  on=['card_id', 'name', 'set_name', 'set_code', 'rarity'], 
                  how='left')

full
```
