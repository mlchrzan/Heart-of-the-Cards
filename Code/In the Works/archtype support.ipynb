```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
import polars as pl
import plotnine as p9
import datetime
```

```python
df = pl.read_csv('../Data/Full Pulls/ygo_full_daily_2025-09-19_19-43-09.csv', null_values="Unknown")
```

```python
df
```

```python
# Get the month and year from tcg_date by using regex to pull the numbers from between YEAR-MONTH-DAY 
df = df.to_pandas()
df['tcg_month'] = df['tcg_date'].str.extract(r'-(\d{1,2})-')[0]
df['tcg_year'] = df['tcg_date'].str.extract(r'(\d{4})-')[0]
# Combine year and month into a new column tcg_month_year in YYYY-MM format and convert to datetime
df['tcg_month_year'] = df['tcg_year'] + '-' + df['tcg_month'].str.zfill(2)
df['tcg_month_year'] = pd.to_datetime(df['tcg_month_year'], format='%Y-%m')

df = pl.from_pandas(df)
df.select(['tcg_date', 'tcg_month', 'tcg_year', 'tcg_month_year'])
```

```python
df_grouped = (df.group_by(['archetype', 'tcg_month_year'])
              .agg(pl.len())
              .sort('tcg_month_year', descending=True)
              .rename({'len': 'count'}))
df_grouped
```

```python
df = df.with_columns([
    pl.col("archetype")
      .map_elements(
          lambda arch: df.filter(pl.col("archetype") == arch)["tcg_month_year"].max(),
          return_dtype=pl.Datetime
      )
      .alias("last_archetype_release")
])
df.select(["archetype", "tcg_month_year", "last_archetype_release"]).head()
```

```python
# Convert to pandas
df_pd = df.to_pandas()

# Calculate time_since_last_release in days
today = pd.Timestamp.today()
df_pd['time_since_last_release'] = (today - df_pd['last_archetype_release']).dt.days

# Convert back to polars
df = pl.from_pandas(df_pd)

df.select(["archetype", "last_archetype_release", "time_since_last_release"]).head()
```

```python
p9.ggplot(df.filter(pl.col('time_since_last_release') > 90), 
          p9.aes(x='time_since_last_release')) + \
    p9.geom_histogram(binwidth=30, fill='blue', alpha=0.7) + \
    p9.labs(title='Distribution of Time Since Last Archetype Release',
            x='Days Since Last Release',
            y='Count') + \
    p9.theme_minimal()
```

```python
(df.sort('time_since_last_release', descending=True)
 .filter(pl.col('time_since_last_release').is_not_null())
 .select(['archetype', 'last_archetype_release', 'time_since_last_release'])
 .to_pandas())
```
